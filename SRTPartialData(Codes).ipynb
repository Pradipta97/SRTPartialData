```python
import numpy as np
from scipy.integrate import quad
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
n=150 #nodal points used to plot f(x)
a=0.001 #left endpoint of the interval (a,b) for the function f(x)
b=1 #right endpoint of the interval (a,b) for the function f(x)
c=0.0001 #left endpoint of the interval of (c,d) for the inversion
d=1 #right endpoint of the interval of (c,d) for the inversion
m=150 #nodal points used in the inversion
def f1(x):
    return (np.exp((-(x-.5)**2)/(2*(.05)**2))/2) #f1 is the given function f(x)
def f(x,t):
  return f1(x)*x
def g(t):
  integral,_=quad(f,1-t,1,args=(t,)) #g(t)=h_0(t)
  return integral
t1=np.linspace(a,b,n)
h=np.zeros(n)
for i in range(n):
  h[i]=g(t1[i])
h1=np.gradient(h,t1) #h1(t)=(h_0)^(1)(t)
def h2(t):
  return np.interp(t,t1,h1) #interpolating the function h1(t)=(h_0)^(1)(t)
def h3(t):
  return h2(1-t) #h3(t)=(h_0)^(1)(1-t)
def h4(t):
  return (h3(t)/t) #using the formula from example (6.1)
t2=np.linspace(c,d,m)
h5=np.zeros(m)
for i in range(m):
  h5[i]=h4(t2[i])
f2=np.zeros(n)
for i in range(n):
  f2[i]=f1(t1[i])
plt.plot(t1,f2,'r',label=r"$f(x)$") #plotting the actual function f(x)
plt.legend(fontsize=13)
plt.xlabel("x-axis ",fontsize=8)
plt.xticks(np.linspace(0,1,10),[f'{tick:.1f}' for tick in np.linspace(0,1,10)])
plt.ylabel("y-axis ",fontsize=8)
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import quad
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import warnings
from scipy.integrate import IntegrationWarning
n=150 #no of points used to plot f(x)
a=0.01 #left endpoint of the interval (a,b) for the function f(x)
b=1 #right endpoint of the interval (a,b) for the function f(x)
c=0.03 #left endpoint of the interval (c,d) for the function obtained from inversion
d=1 #left endpoint of the interval (c,d) for the function obtained from inversion
m=150 #no of points used in the inversion
def f1(x):
    return (np.exp((-(x-.5)**2)/(2*(.05)**2))/2) #f1 is the given function f(x)
def f(x,t):
  return f1(x)*x*((1+t)**2-x**2)*(x**2-(1-t)**2)
def g(t):
  integral,_=quad(f,1-t,1,args=(t,))  #g(t)=h_1(t)
  return integral
t1=np.linspace(a,b,n)
h=np.zeros(n)
for i in range(n):
  h[i]=g(t1[i])
h1=np.gradient(h,t1)  #h1(t)=(h_1)^(1)(t)
h2=np.gradient(h1,t1) #h2(t)=(h_1)^(2)(t)
h3=np.zeros(n)
for i in range(n):
  h3[i]=-h1[i]/(t1[i]**3)+h2[i]/(t1[i]**2) #to compute D^(2)h_1(t) we use the formula in our paper (lemma 4.3) so,h3=D^(2)h_1(t)
h4=np.gradient(h3,t1) #h4(t)=(D^(2)h_1)^(1)(t)
def h6(t):
   return np.interp(t,t1,h4)  #interpolating the function h4(t)=(D^(2)h_1)^(1)(t)
def h7(t):
  return (np.exp(t)*(1-t)*h6(t)) #integrand in the explicit formula (example 6.2)
def h8(y):
    integral,_=quad(h7,c,y) #integrating the integrand from c to y
    return integral
def h9(y):
  return h8(1-y)
def h10(y):
  return ((1-y)*np.exp(y-1)*h9(y))/(8*y**3) #the explicit formula from (example 6.2)
f2=np.zeros(n)
for i in range(n):
  f2[i]=f1(t1[i])
y=np.linspace(c,d,m)
h11=np.zeros(m)
for i in range(m):
  h11[i]=h10(y[i])
warnings.filterwarnings("ignore", category=IntegrationWarning)
plt.plot(t1,f2,'r',label="f(x)") #plotting the actual function f(x)
plt.legend(fontsize=13)
plt.xlabel("x-axis ",fontsize=8)
plt.xticks(np.linspace(0,1,10),[f'{tick:.1f}' for tick in np.linspace(0,1,10)])
plt.ylabel("y-axis ",fontsize=8)
plt.grid(True)
plt.show()
```

7. If we want to recover $z(t)=f(t)t^3$ using the ODE for $z(t)$, then we can reconstruct $z(t)$ very close to $0$ also.

```python
import numpy as np
from scipy.integrate import quad
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import warnings
from scipy.integrate import IntegrationWarning
n=300 #no of points used to plot f(x)
a=0.01 #left endpoint of the interval (a,b) for the function f(x)
b=0.99 #right endpoint of the interval (a,b) for the function f(x)
c=0.000000001 #left endpoint of the interval of (c,d) for the ODE
d=0.99 #right endpoint of the interval of (c,d) for the ODE
c1=0.000000001 #left endpoint of the interval of (c,d) for the evaluation of ODE
d1=0.99 #right endpoint of the interval of (c,d) for the evaluation of ODE
m=150 #no of points used in the inversion
def f1(x):
    if 1/4<x<1/2:
       return 4*x-1
    elif 1/2<=x<3/4:
       return 3-4*x
    return 0    #f1 is the given function f(x)
def f(x,t):
  return f1(x)*x*((1+t)**2-x**2)*(x**2-(1-t)**2)
def g(t):
  integral,_=quad(f,1-t,1,args=(t,)) #g(t)=h_1(t),
  return integral
t1=np.linspace(a,b,n)
h=np.zeros(n)
for i in range(n):
  h[i]=g(t1[i])
h1=np.gradient(h,t1) #h1(t)=(h_1)^(1)(t)
h2=np.gradient(h1,t1) #h2(t)=(h_1)^(2)(t)
h3=np.zeros(n)
for i in range(n):
  h3[i]=-h1[i]/(t1[i]**3)+h2[i]/(t1[i]**2) #to compute D^(2)h_1(t) we use the formula in our paper (lemma 4.3) so,h3=D^(2)h_1(t)
h4=np.gradient(h3,t1)  #h4(t)=(D^(2)h_1)^(1)(t)
def f5(t):
  return np.interp(t,t1,h4) #interpolating the function h4(t)=(D^(2)h_1)^(1)(t)
def f6(t):
  return f5(1-t) #f6(t)=(D^(2)h_1)^(1)(1-t)
def odefn(t,y):
  return (-1*t*(1-t)*f6(t))/(8)+(-(t)*y)/(1-t) #from the ode in example (6.2), substituting z(t)=f(t)*t**3, we have ODE interms of z(t), which is z'(t)=(t*(1-t)*(D^(2)h_1)^(1)(1-t))/(-8)-(t*z(t))/(1-t).
t2=np.linspace(d1,c1,m)
sol = solve_ivp(odefn,(d,c),[0],method='RK45',t_eval=t2) #ODE solver to solve the ode
f2=np.zeros(n)
for i in range(n):
  f2[i]=f1(t1[i])*(t1[i])**3
warnings.filterwarnings("ignore", category=IntegrationWarning)
plt.plot(t1,f2,'r',label="f(x)") #plotting the actual function f(x)
plt.legend(fontsize=13)
plt.xlabel("x-axis ",fontsize=8)
plt.xticks(np.linspace(0,1,10),[f'{tick:.1f}' for tick in np.linspace(0,1,10)])
plt.ylabel("y-axis ",fontsize=8)
plt.grid(True)
plt.show()
```

```python
plt.plot(t2,sol.y[0],'b',label="From inversion ") #plotting the function which is obtained from inversion
plt.legend(fontsize=10)
plt.xlabel("x-axis ",fontsize=8)
plt.xticks(np.linspace(0,1,10),[f'{tick:.1f}' for tick in np.linspace(0,1,10)])
plt.ylabel("y-axis ",fontsize=8)
plt.grid(True)
plt.show()
```

General function for dimension n=3, Consider $f(x)=f_{0,1}(t)Y_{0,1}(\theta)+f_{1,2}(t)Y_{1,2}(\theta)$, where $x=t\theta$,  $f_{0,1}(t)=\cos(50t)$, $f_{1,2}(t)=\frac{1}{2}e^\frac{{-(x-.7)^2}}{2(.05)^2}$.

```python
import numpy as np
from scipy.integrate import quad
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
n=300
a=0.01 #left endpoint of the interval (a,b) for the function f_{0,1}(x) and f_{1,2}(x)
b=0.99 #right endpoint of the interval (a,b) for the function f_{0,1}(x) and f_{1,2}(x)
c=0.1 #left endpoint of the interval of (c,d) for the ODE
d=0.95 #right endpoint of the interval of (c,d) for the ODE
c1=0.01 #left endpoint of the interval of (c,d) for the inversion
d1=0.99 #right endpoint of the interval of (c,d) for the inversion
c2=0.1 #left endpoint of the interval of (c,d) for the evaluation of ODE
d2=0.95 #right endpoint of the interval of (c,d) for the evaluation of ODE
n1=100 #no of points used in the inversion to recover f_{0,1}(x)
n2=150 #no of points used in the inversion to recover f_{1,2}(x)
def f0(x):
    return (np.cos(30*x)) #f0 is the given function f_{0,1}(x)
def f1(x):
     return (np.exp((-(x-.6)**2)/(2*(.05)**2))/2) ##f1 is the given function f_{1,2}(x)
def f00(x,t):
  return f0(x)*x
def f01(x,t):
  return f1(x)*(1+x**2-t**2)*(1/2)
def g0(t):
  integral,_=quad(f00,1-t,1,args=(t,)) #g0(t)=h_{0,1}(t)
  return integral
def g1(t):
  integral,_=quad(f01,1-t,1,args=(t,)) #g1(t)=h_{1,2}(t)
  return integral
t1=np.linspace(a,b,n)
m=np.zeros(n)
for i in range(n):
  m[i]=g0(t1[i])
m1=np.gradient(m,t1) #m1(t)=(h_{0,1})^(1)(t)
def m2(t):
  return np.interp(t,t1,m1)  #interpolating the function m1(t)=(h_{0,1})^(1)(t)
def m3(t):
  return m2(1-t)  #m3(t)=(h_{0,1})^(1)(1-t)
def m4(t):
  return (m3(t)/t)  #using the formula from example (6.1)
m5=np.zeros(n1)
t3=np.linspace(c1,d1,n1)
for i in range(n1):
  m5[i]=m4(t3[i])
f2=np.zeros(n)
for i in range(n):
  f2[i]=f0(t1[i])
h=np.zeros(n)
for i in range(n):
  h[i]=g1(t1[i])
h1=np.gradient(h,t1)  #h1(t)=(h_{1,2})^(1)(t)
h2=np.zeros(n)
for i in range(n):
  h2[i]=h1[i]/t1[i]
h3=np.gradient(h2,t1) #h3(t)=(D^(1)h_{1,2})^(1)(t)
def h4(t):
  return np.interp(t,t1,h3) #interpolating the function h3(t)=(D^(1)h_{1,2})^(1)(t)
def h5(t):
  return h4(1-t)  #h5(t)=(D^(1)h_{1,2})^(1)(1-t)
def odefn(t,y):
  return -((1-t)*h5(t)/t**2)-((2-t+(1-t)**2)*y)/(t*(1-t))  #from the ode for general functions
t2=np.linspace(d2,c2,n2)
sol = solve_ivp(odefn,(d,c),[0],method='RK45',t_eval=t2)  #ODE solver to solve the ode
f4=np.zeros(len(t2))
for i in range(len(t2)):
  f4[i]=sol.y[0][i]*t2[i] #recovering the function f_{1,2}(t)
f3=np.zeros(n)
for i in range(n):
  f3[i]=f1(t1[i])
plt.plot(t1, f2, 'r', label=r"$f_{_{0,1}}(t)$")  #plotting the actual function f_{0,1}(x)
plt.legend(fontsize=9)
plt.xticks(np.linspace(0,1,10),[f'{tick:.1f}' for tick in np.linspace(0,1,10)])
plt.xlabel("x-axis ",fontsize=8)
plt.ylabel("y-axis ",fontsize=8)
plt.grid(True)
plt.show()
```
